## Importing Libraries

```python
#Importing matplotlib library for visualization.
import cv2
import matplotlib.pyplot as plt
```

## Importing required Image


```python
img=cv2.imread("/content/images.jpg")

#Displaying the image.
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

##Viewing Image into RGB format

```python
#OpenCV uses BGR color scheme whereas matplotlib uses RGB colors scheme.
#Converting BGR into RGB
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
plt.savefig('temp.png')
plt.show()
```

## Converting Image into Grayscale Image

```python
grey_image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) 
from google.colab.patches import cv2_imshow
cv2_imshow(grey_image)
```

## Inverting the grayscale image

```python
inverted_image = cv2.bitwise_not(grey_image)  
from google.colab.patches import cv2_imshow
cv2_imshow(inverted_image)
```

## Bluring the inverted image

```python
image_blur = cv2.GaussianBlur(inverted_image, (21, 21),sigmaX=0, sigmaY=0)
from google.colab.patches import cv2_imshow
cv2_imshow(image_blur)
```

## Pencil Sketching the Image

```python
def pencil_sketch(x, y):

    return cv2.divide(x, 255 - y, scale=256)
```

```python
pencil_sketch=pencil_sketch(grey_image,image_blur)
from google.colab.patches import cv2_imshow
cv2_imshow(pencil_sketch)
```

## Comparing the Original and Sketched Image

```python
plt.figure(figsize=(14,8))

plt.subplot(1,2,1)
plt.title('Original image of the Family', size=18)
plt.imshow(RGB_img)
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Sketch image of the Family', size=18)
rgb_sketch=cv2.cvtColor(pencil_sketch, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.savefig('t1.png')
plt.show()
```
